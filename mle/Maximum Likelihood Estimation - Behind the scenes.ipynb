### Distância entre duas funções

O espaço vetorial de funções reais cujo domínio é o intervalo \[a,b\] é um espaço com [produto interno](https://mathworld.wolfram.com/InnerProduct.html):

$$\large <f, g> = \int_{a}^{b} f(x) g(x) \,dx $$

Todo espaço com produto interno é um espaço métrico, cuja distância (L2) é dada por:

$$\large  d(f,g) = \lVert f - g \rVert = <f - g, f - g>^{\frac{1}{2}} = \sqrt{\int_{a}^{b} (f(x) - g(x))^{2} \,dx} $$

Como exemplo, a distância entre o seno e cosseno no intervalo \[0,2$\pi$\] é dada como:

$$\large  d(\sin{\theta},\cos{\theta}) = \sqrt{\int_{0}^{2\pi} (\sin{\theta} - \cos{\theta})^{2} \,d\theta} = \sqrt{2\pi} $$
<br>
$$\large  d(\sin{\theta},\sin{\theta}) = 0 $$

#### Propriedades

Seguem as propriedades de uma [métrica](https://mathworld.wolfram.com/Metric.html):

1. Simetria: $d(x,y) = d(y,x)$
2. Desigualdade Triangular: $d(x,y) + d(y,z) \geq d(x,z)$
3. Sempre positiva: $d(x,y) \geq 0$
4. Se $d(x,y) = 0$ então $x = y$

### Distância entre distribuições de probabilidade: Total Variation Distance

Na estatística, uma métrica utilizada para medir distância entre duas distribuições é a [Total Variation Distance](https://en.wikipedia.org/wiki/Total_variation_distance_of_probability_measures), dada por:

$$\large  TV(P_{\theta^{*}},P_{\theta}) = \max_{A \in E}{|P_{\theta^{*}} - P_{\theta}|} = \frac{1}{2}\int_{E} |\rho_{\theta^{*}}(x) - \rho_{\theta}(x)| \,dx$$

Há conexão direta entre TV e a norma L1, além disso, supondo que $P_{\theta}$ é contínua, na ingetral está presente a densidade de probabilidade associada à distribuição. 

<center><strong>Inútil!!! Para calcular essa distância, preciso conhecer $ P_{\theta^{*}}$, porém se conhecer $ P_{\theta^{*}}$ não preciso do modelo.</strong></center>

### Divergência de Kullback-Leibler

Esta quantidade $D_{KL}$ não é uma métrica, pois não obedece as propriedades da simetria nem desigualdade triangular, porém satisfaz as outras duas condições remanescentes. Assim a minimização de $D_{KL}(P_{\theta^{*}},P_{\theta})$ acarreta que $\theta^{*}=\theta$ e, portanto, $P_{\theta^{*}}=P_{\theta}$. A divergência de Kullback-Leibler também é conhecida como entropia relativa.

**No contexto de machine learning também é conhecido como *information gain*.** 

A divergência $D_{KL}(P_{\theta^{*}},P_{\theta})$ pode ser calculada da seguinte fórmula:

$$\large D_{KL}(P_{\theta^{*}},P_{\theta}) = \int \rho_{\theta^{*}}(x)\ln{\frac{\rho_{\theta^{*}}(x)}{\rho_{\theta}(x)}} \,dx$$

Porém agora é possível estimar a divergência usando:

$$\large D_{KL}(P_{\theta^{*}},P_{\theta}) = \sum p_{\theta^{*}}(x)\ln{\frac{p_{\theta^{*}}(x)}{p_{\theta}(x)}} \,dx = \text{constante} - \sum p_{\theta^{*}}(x)\ln{p_{\theta}(x)} =  \text{constante} + \mathbb{E}_{\theta^{*}}[{\ln{p_{\theta}(x)}}]$$
<br>
Pela lei dos grandes números, pode-se construir um estimator para a esperança, o mesmo utilizado em métodos de [Monte Carlo](https://ib.berkeley.edu/labs/slatkin/eriq/classes/guest_lect/mc_lecture_notes.pdf):
<br>

$$\large \mathbb{E}_{\theta^{*}}[{\ln{p_{\theta}(x)}}] 	\rightsquigarrow \frac{1}{n} \sum \ln{p_{\theta}(x_{i})}$$ 
<br>
<center><strong>Não precisamos conhecer $P_{\theta^{*}}$ para construir um estimador para a divergência KL!!!</strong></center>

Assim:
$$\large \hat{D}_{KL}(P_{\theta^{*}},P_{\theta}) = \text{constante} - \frac{1}{n} \sum_{i=0}^{n} \ln{p_{\theta}(x_{i})}$$

Lembrando que o objetivo é minimizar a divergência KL, portanto:

$$\large \min_{\theta \in \Theta}{\hat{D}_{KL}(P_{\theta^{*}},P_{\theta})} \Leftrightarrow  \max_{\theta \in \Theta}{\frac{1}{n} \sum_{i=0}^{n} \ln{p_{\theta}(x_{i})}} \Leftrightarrow  \max_{\theta \in \Theta}{ \prod_{i=0}^{n} p_{\theta}(x_{i})} \Leftrightarrow  \max_{\theta \in \Theta}{ p_{\theta}(\vec{x})} $$ 
<br>

<center><strong>Ou seja minimizando a divergência KL estamos, maximização da verossimilhança!!!</strong></center>

A verossimilhança também é comumemte escrita como:

$$\large L(\theta|\vec{x}) = p_{\theta}(\vec{x})$$
