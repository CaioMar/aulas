### Distncia entre duas funes

O espao vetorial de funes reais cujo domnio  o intervalo \[a,b\]  um espao com [produto interno](https://mathworld.wolfram.com/InnerProduct.html):

$$\large <f, g> = \int_{a}^{b} f(x) g(x) \,dx $$

Todo espao com produto interno  um espao mtrico, cuja distncia (L2)  dada por:

$$\large  d(f,g) = \lVert f - g \rVert = <f - g, f - g>^{\frac{1}{2}} = \sqrt{\int_{a}^{b} (f(x) - g(x))^{2} \,dx} $$

Como exemplo, a distncia entre o seno e cosseno no intervalo \[0,2$\pi$\]  dada como:

$$\large  d(\sin{\theta},\cos{\theta}) = \sqrt{\int_{0}^{2\pi} (\sin{\theta} - \cos{\theta})^{2} \,d\theta} = \sqrt{2\pi} $$
<br>
$$\large  d(\sin{\theta},\sin{\theta}) = 0 $$

#### Propriedades

Seguem as propriedades de uma [mtrica](https://mathworld.wolfram.com/Metric.html):

1. Simetria: $d(x,y) = d(y,x)$
2. Desigualdade Triangular: $d(x,y) + d(y,z) \geq d(x,z)$
3. Sempre positiva: $d(x,y) \geq 0$
4. Se $d(x,y) = 0$ ento $x = y$

### Distncia entre distribuies de probabilidade: Total Variation Distance

Na estatstica, uma mtrica utilizada para medir distncia entre duas distribuies  a [Total Variation Distance](https://en.wikipedia.org/wiki/Total_variation_distance_of_probability_measures), dada por:

$$\large  TV(P_{\theta^{*}},P_{\theta}) = \max_{A \in E}{|P_{\theta^{*}} - P_{\theta}|} = \frac{1}{2}\int_{E} |\rho_{\theta^{*}}(x) - \rho_{\theta}(x)| \,dx$$

H conexo direta entre TV e a norma L1, alm disso, supondo que $P_{\theta}$  contnua, na ingetral est presente a densidade de probabilidade associada  distribuio. 

<center><strong>Intil!!! Para calcular essa distncia, preciso conhecer $ P_{\theta^{*}}$, porm se conhecer $ P_{\theta^{*}}$ no preciso do modelo.</strong></center>

### Divergncia de Kullback-Leibler

Esta quantidade $D_{KL}$ no  uma mtrica, pois no obedece as propriedades da simetria nem desigualdade triangular, porm satisfaz as outras duas condies remanescentes. Assim a minimizao de $D_{KL}(P_{\theta^{*}},P_{\theta})$ acarreta que $\theta^{*}=\theta$ e, portanto, $P_{\theta^{*}}=P_{\theta}$. A divergncia de Kullback-Leibler tambm  conhecida como entropia relativa.

**No contexto de machine learning tambm  conhecido como *information gain*.** 

A divergncia $D_{KL}(P_{\theta^{*}},P_{\theta})$ pode ser calculada da seguinte frmula:

$$\large D_{KL}(P_{\theta^{*}},P_{\theta}) = \int \rho_{\theta^{*}}(x)\ln{\frac{\rho_{\theta^{*}}(x)}{\rho_{\theta}(x)}} \,dx$$

Porm agora  possvel estimar a divergncia usando:

$$\large D_{KL}(P_{\theta^{*}},P_{\theta}) = \sum p_{\theta^{*}}(x)\ln{\frac{p_{\theta^{*}}(x)}{p_{\theta}(x)}} \,dx = \text{constante} - \sum p_{\theta^{*}}(x)\ln{p_{\theta}(x)} =  \text{constante} + \mathbb{E}_{\theta^{*}}[{\ln{p_{\theta}(x)}}]$$
<br>
Pela lei dos grandes nmeros, pode-se construir um estimator para a esperana, o mesmo utilizado em mtodos de [Monte Carlo](https://ib.berkeley.edu/labs/slatkin/eriq/classes/guest_lect/mc_lecture_notes.pdf):
<br>

$$\large \mathbb{E}_{\theta^{*}}[{\ln{p_{\theta}(x)}}] 	\rightsquigarrow \frac{1}{n} \sum \ln{p_{\theta}(x_{i})}$$ 
<br>
<center><strong>No precisamos conhecer $P_{\theta^{*}}$ para construir um estimador para a divergncia KL!!!</strong></center>

Assim:
$$\large \hat{D}_{KL}(P_{\theta^{*}},P_{\theta}) = \text{constante} - \frac{1}{n} \sum_{i=0}^{n} \ln{p_{\theta}(x_{i})}$$

Lembrando que o objetivo  minimizar a divergncia KL, portanto:

$$\large \min_{\theta \in \Theta}{\hat{D}_{KL}(P_{\theta^{*}},P_{\theta})} \Leftrightarrow  \max_{\theta \in \Theta}{\frac{1}{n} \sum_{i=0}^{n} \ln{p_{\theta}(x_{i})}} \Leftrightarrow  \max_{\theta \in \Theta}{ \prod_{i=0}^{n} p_{\theta}(x_{i})} \Leftrightarrow  \max_{\theta \in \Theta}{ p_{\theta}(\vec{x})} $$ 
<br>

<center><strong>Ou seja minimizando a divergncia KL estamos, maximizao da verossimilhana!!!</strong></center>

A verossimilhana tambm  comumemte escrita como:

$$\large L(\theta|\vec{x}) = p_{\theta}(\vec{x})$$
